####  matplotlib - plotting library for python

#### seaborn - data visualization library based on matplotlib

```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import random
```

##### line plot

```python
names = ['a','b','c','d','e']
actual_weight = [56,48,61,67,50]

plt.figure(figsize=(4,4))
plt.plot(names,actual_weight)
plt.xlabel('Names')
plt.ylabel('Actual weight')
plt.title('Weight Diagram')
```

```python
actual_weight = [56,48,61,67,50]
desired_weight = [47, 50,65,68,54]
names = ['a','b','c','d','e']
plt.plot(names, actual_weight, color = 'brown', label='actual_weight')
plt.plot(names,desired_weight, color = 'pink', label='desired_weight')

plt.xlabel('Names')
plt.ylabel('weight')
plt.title('Multiple line plot',{'fontsize':15, 'color':'red'})
plt.legend()
```

```python
actual_weight = [56,48,61,67,50]
desired_weight = [47, 50,65,68,54]
names = ['a','b','c','d','e']

plt.plot(names, actual_weight, color = 'brown', label='actual_weight')
plt.xlabel('Names')
plt.ylabel('weight')
plt.figure()

plt.plot(names,desired_weight, color = 'pink', label='desired_weight')
plt.xlabel('Names')
plt.ylabel('weight')
```

##### Box plots (Whisker plot)
Box plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

It is useful for detecting outliers.

```python
data1 = pd.read_csv('healthcare-dataset-stroke-data.csv')
data1.head()
```

```python
plt.boxplot(data1['age'])
```

```python
plt.boxplot(data1['age'], patch_artist=True)
```

```python
plt.boxplot(data1['avg_glucose_level'], patch_artist=True)
```
