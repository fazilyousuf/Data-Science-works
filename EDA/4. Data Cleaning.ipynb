#### Handling missing values

```python
# checking for null values first
data.isnull().sum()
```

##### Removing missing values

```python
data_copy = data.dropna()
data_copy
```

```python
data_copy.isnull().sum()
```

##### Filling missing values using SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy ='most_frequent' ) # default strategy= 'mean'
```

```python
data_copy2 = data.copy()
data_copy2['bmi'] = imputer.fit_transform(data[['bmi']])
```

```python
data_copy2.isnull().sum()
```

##### filling missing values using fillna

```python
data_copy3 = data.copy()
bmi_mean = data['bmi'].mean()
data_copy3['bmi'] = data['bmi'].fillna(bmi_mean)
data_copy3.isnull().sum()
```

##### Removing outliers using interquartile range (IQR)
Any data point outside the range of 1.5 times the IQR below the 25th percentile or above the 75th percentile can be considered an outlier.

```python
from matplotlib import pyplot as plt
```

```python
plt.boxplot(data['avg_glucose_level'], patch_artist=True)
```

```python
import numpy as np
print(f'Old shape of data: {data.shape}')

Q1 = data.avg_glucose_level.quantile(0.25)
Q3 = data.avg_glucose_level.quantile(0.75)
IQR = Q3-Q1
print(Q1, Q3, IQR)

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_index = np.where(data['avg_glucose_level'] >= upper)[0]
lower_index = np.where(data['avg_glucose_level'] <= lower)[0]

data.drop(index = upper_index, inplace = True)
data.drop(index = lower_index, inplace = True)

print(f'New shape of data: {data.shape}')
```
