##### Getting descriptive stats of the dataset

```python
count = data['sepal.length'].count()
print(f'Count: {count}')

mean = data['sepal.length'].mean()
print(f'Mean: {mean}')

median = data['sepal.length'].median()
print(f'Median: {median}')

var = data['sepal.length'].var()
print(f'Variance: {var}')

std = data['sepal.length'].std()
print(f'Standard deviation: {std}')

minimum = data['sepal.length'].min()
print(f'Minimum: {minimum}')

q1 = data['sepal.length'].quantile(q=0.25)
print(f'25%: {q1}')

q2 = data['sepal.length'].quantile(q=0.50)
print(f'50% (Median): {q2}')

q3 = data['sepal.length'].quantile(q=0.75)
print(f'75%: {q3}')

maximum = data['sepal.length'].max()
print(f'Maximum: {maximum}')
```
